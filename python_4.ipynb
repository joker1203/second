```python
import time
import calendar
k=0
def nowtime():
    t1=time.time()
    localt1=time.localtime(t1)
    localtime=time.asctime(localt1)
    print(localtime)
while k<=1:
    k+=1
    time.sleep(2)
    nowtime()
for i in range(9,13):
    print(i,calendar.month(2019,i))
    print('----------------------')
for i in range(1,8):
    print(i,calendar.month(2020,i))
    print('----------------------')
    time.sleep(1)
```
